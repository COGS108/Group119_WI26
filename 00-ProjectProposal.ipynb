## Research Question

Algorithm-driven social media platforms like TikTok and Instagram Reels have become central to college students' daily routines, yet their impact on cognitive and academic outcomes remains unclear. What is the correlation between daily time spent on algorithm-driven social media platforms (TikTok, Instagram Reels, YouTube Shorts) and college students' self-reported attention span, mood state, and academic productivity? Additionally, do these correlations vary significantly across different global regions? This statistical inference approach will measure usage in hours per day and assess outcomes through self-report scales.

## Background and Prior Work

Algorithm-driven social media platforms such as TikTok and Instagram Reels play a major role and heavily influence the daily lives of college students, and for many, scrolling through these apps has become a routine part of the day. These platforms rely on personalized algorithms that continuously adapt to user behavior in order to deliver content that maximizes engagement.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Because this design makes social media highly engaging and tailored to individual interests, it has raised concerns as to how constant exposure to short-form, algorithm-based content may affect attention, mood, and academic productivity, particularly amongst Gen Z college students.

Prior research suggests that extensive social media use can have meaningful psychological and academic consequences. A mixed-methods study by El-Shihy, which examined Gen Z students in Egypt, found that higher levels of social media addiction were associated with increased stress, reduced concentration, and lower academic engagement.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Many students in the study reported difficulty maintaining focus while studying and challenges with time management due to frequent social media use. These findings indicate that students are often aware of the negative effects of excessive social media use, especially when it interferes with their ability to concentrate on academic tasks.

Other research highlights the specific role of algorithms, rather than social media use in general. In Pan's study on TikTok's personalized algorithm, it was evident that college students reported spending more time on the platform because the content felt highly relevant and difficult to disengage from.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) When asked about their tangible experiences, many shared that they were losing track of time, procrastinating more, and experienced a reduction in maintaining their focus and attention during tasks that typically require sustained engagement. Similarly, Gamboa et al. examined the long-term effects of algorithm-driven content consumption on youth populations between the ages of 13-21 and found associations related to changes in attention patterns, mood regulation, and psychological perceptions over time.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Together, these studies suggest that algorithm-driven platforms may influence how students allocate attention and manage academic responsibilities. However, much of the existing research focuses on specific platforms or regions, leaving a gap in understanding how these relationships may vary across global contexts. Our project builds on this already relevant research by further examining how time spent on algorithm-driven social media relates to college students' self-reported attention, mood, and academic productivity across different regions of the world.

## Hypothesis


We hypothesize that time spent on algorithm-driven social media platforms will show a moderate to strong negative correlation with college students’ self-reported attention and academic productivity, such that increased usage is associated with decreased attentional control and lower productivity. (2) Additionally, we predict that more time spent on these platforms will demonstrate a moderate positive correlation with negative moods (e.g., stress, anxiety, or low affect). (3) The magnitude of these relationships are expected to vary across global regions, this is due to the varying cultural differences in technology use and differences in academic expectations across the globe. 

## Data

To answer our research question, we would use a survey dataset collected from college students across different regions of the world. Each response in the dataset would represent one student. We would include basic demographic information such as age, gender, country or region, year in school, and major to better understand how social media use may differ across groups.

The main independent variables would focus on social media usage, including how much time students spend per day on algorithm-driven platforms like TikTok, Instagram Reels, and YouTube Shorts. This could be measured using numerical values (such as hours per day) or categories (such as “less than 1 hour,” “1–3 hours,” or “more than 3 hours”). We would also include which platform the student uses most often.

The main dependent variables would measure students’ self-reported attention, mood, and academic productivity. These could include questions about how well they feel they can focus, their overall mood or stress level, and how productive they feel academically. These would likely be measured using rating-scale questions (for example, from “strongly disagree” to “strongly agree”). We could also include GPA or hours spent studying per week as additional indicators of academic productivity.

The dataset would include several hundred students so that we could compare patterns across different regions. The data would be cross-sectional, meaning it is collected at one point in time.

A dataset we found that could be useful is the Social Media Addiction vs Relationships dataset on Kaggle ([https://www.kaggle.com/datasets/adilshamim8/social-media-addiction-vs-relationships](https://www.kaggle.com/datasets/adilshamim8/social-media-addiction-vs-relationships)).

This dataset is publicly available and can be downloaded directly from Kaggle with a free account. The data comes from survey responses and focuses on social media usage habits and how they relate to behavior and personal relationships. Important variables in this dataset include how much time respondents spend on social media, indicators of social media addiction, and self-reported impacts on daily life. While the dataset is not specifically limited to college students or attention and academics, some of the usage and behavioral variables could still be relevant to our project. We would likely need to filter or adapt the data to focus on younger respondents and variables that relate more closely to attention, mood, or productivity.

 The dataset is susceptible to representation bias and selection bias, particularly regarding the “digital divide.” The data we collect inherently excludes students who may not have access to high-speed internet or high-end smartphones capable of running these apps. To ensure an equitable analysis, we must acknowledge that social media algorithms themselves reflect algorithmic bias, where certain content or demographics are prioritized over others. If we don’t carefully balance our dataset, the conclusions might suggest that “average” productivity looks like a specific demographic's experience, further marginalizing students from different socioeconomic or cultural backgrounds.

 Before beginning the analysis, we will conduct a demographic audit to compare our sample’s diversity against global university enrollment statistics. We will use stratified analysis to look for patterns within specific subgroups rather than just looking at the aggregate “average student,” which often hides disparate impacts. When communicating results, we will explicitly state the limitations of our sample and avoid making universal claims if certain populations are underrepresented.
